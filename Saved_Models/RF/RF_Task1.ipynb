```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import hamming_loss
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
```

```python
data2=pd.read_csv('/home/user/Downloads/test_emb(1).csv')
print(data2)
```

```python
data1=pd.read_csv('/home/user/Downloads/train_emb(1).csv')
print(data1)
```

```python
data1.drop(data1.tail(1).index,inplace=True)
```

```python
print(data1)
```

```python
train_X = data1[data1.columns[0:100]]
train_Y = data1[data1.columns[100:]]
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(train_X,train_Y)
```

```python
test_X = data2[data2.columns[0:100]]
test_Y = data2[data2.columns[100:]]
y_pred=clf.predict(test_X)
print("Accuracy:",(metrics.accuracy_score(test_Y, y_pred)*100))
print("Log Loss:",log_loss(test_Y, y_pred))
print("Hamming Loss:",hamming_loss(test_Y, y_pred))
```

```python
import pickle
filename = 'RF_Task1_TrainedModel.pkl'
pickle.dump(clf, open(filename, 'wb'))
```
